```python
import pandas as pd
```

```python
df = pd.read_csv("Mumbai House Prices.csv")
```

```python
df.head()
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
df['type'].value_counts()
```

```python
df['type'] = df['type'].replace(['Independent House','Penthouse'],'other')
```

```python
df.head()
```

```python
df['locality'].value_counts()
```

```python
top_localities = df['locality'].value_counts().nlargest(20).index
df['locality'] = df['locality'].where(df['locality'].isin(top_localities),'other')
```

```python
df['area'].value_counts()
```

```python
df['area'].describe()
```

```python
df['price'].value_counts()
```

```python
df['price'] = df.apply(lambda x:x['price']*100 if x['price_unit'] == 'Cr'else x['price'],axis=1)
```

```python
df['price'] = df['price']*100000
```

```python
df.head()
```

```python
df.drop(columns='price_unit',inplace=True)
```

```python
df.head()
```

```python
df['region'].value_counts()
```

```python
top_region = df['region'].value_counts().nlargest(20).index
df['region'] = df['region'].where(df['region'].isin(top_region),'other')
```

```python
df.head()
```

```python
df['status'].value_counts()
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
df['status'] = le.fit_transform(df['status'])
```

```python
df.head()
```

```python
df['age'].value_counts()
```

```python
df.head()
```

```python
df.to_csv("cleaned_data.csv",index=None)
```
